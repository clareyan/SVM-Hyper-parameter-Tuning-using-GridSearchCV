```python
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
import pandas as pd
# Loading some example data
iris = datasets.load_iris()
```

```python
# Import data from the external CSV
irisdata = pd.read_csv('iris.csv')
```

```python
irisdata.head()
```

```python
# select two columns 
X=irisdata[['sepal_length', 'sepal_width']].to_numpy() 
y=irisdata[['class']].to_numpy()
```

```python
X = iris.data[:, :2] # we only take the first two features.
y = iris.target
```

```python
# Training a classifier
svm = SVC(C=0.5, kernel='linear')
svm.fit(X, y)

# Plotting decision regions
plot_decision_regions(X, y, clf=svm, legend=2)

# Adding axes annotations
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.title('SVM on Iris_linear')
plt.show()
```

```python
# Training a classifier
svm = SVC(C=0.5, kernel='rbf')
svm.fit(X, y)

# Plotting decision regions
plot_decision_regions(X, y, clf=svm, legend=2)

# Adding axes annotations
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.title('SVM on Iris_rbf')
plt.show()
```

```python
# Training a classifier
svm = SVC(C=0.5, kernel='sigmoid')
svm.fit(X, y)

# Plotting decision regions
plot_decision_regions(X, y, clf=svm, legend=2)

# Adding axes annotations
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.title('SVM on Iris_sigmiod')
plt.show()
```
